# 1) Suma de señales continuas

El siguiente códido genera dos señales senoidales continuas de diferente frecuencia y amplitud.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la señal 1
amplitud_1 = 1.5
frecuencia_1 = 2.0
fase_1 = 0.0

# Parámetros de la señal 2
amplitud_2 = 0.5
frecuencia_2 = 75
fase_2 = np.pi / 2

# Parámetros de tiempo
tiempo_inicial = 0.0
tiempo_final = 2.0
puntos = 1000

# Generar puntos de tiempo
tiempo = np.linspace(tiempo_inicial, tiempo_final, puntos)

# Generar señal 1
señal_1 = amplitud_1 * np.sin(2 * np.pi * frecuencia_1 * tiempo + fase_1)

# Generar señal 2
señal_2 = amplitud_2 * np.sin(2 * np.pi * frecuencia_2 * tiempo + fase_2)

# Sumar señales
suma_señales = señal_1 + señal_2

# Graficar señal 1
plt.subplot(3, 1, 1)
plt.plot(tiempo, señal_1)
plt.title('Señal 1')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')

# Graficar señal 2
plt.subplot(3, 1, 2)
plt.plot(tiempo, señal_2)
plt.title('Señal 2')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')

# Graficar suma de señales
plt.subplot(3, 1, 3)
plt.plot(tiempo, suma_señales)
plt.title('Suma de señales')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')

# Ajustar espaciado entre subplots
plt.tight_layout()

# Mostrar la gráfica
plt.show()


```

# 2) Suma de señales discretas

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de las señales
frecuencia = 1  # Frecuencia de las señales en Hz
muestras_por_segundo = 10  # Número de muestras por segundo
duracion = 2  # Duración de las señales en segundos
fase1 = 3*np.pi/2  # Fase de la primera señal en radianes
fase2 = np.pi/2  # Fase de la segunda señal en radianes

# Generar el tiempo discreto de muestreo
muestras_totales = int(muestras_por_segundo * duracion)
t = np.arange(muestras_totales)

# Generar las señales senoidales
senal1 = np.sin(2 * np.pi * frecuencia * t / muestras_por_segundo + fase1)
senal2 = np.sin(2 * np.pi * frecuencia * t / muestras_por_segundo + fase2)

# Sumar las señales
senal_sumada = senal1 + senal2

# Graficar las señales
plt.figure(figsize=(10, 6))

plt.subplot(3, 1, 1)
plt.stem(t, senal1, label='Señal 1', use_line_collection=True)
plt.title('Señal 1')
plt.legend()

plt.subplot(3, 1, 2)
plt.stem(t, senal2, label='Señal 2', use_line_collection=True)
plt.title('Señal 2')
plt.legend()

plt.subplot(3, 1, 3)
plt.stem(t, senal_sumada, label='Suma de señales', use_line_collection=True)
plt.title('Suma de señales')
plt.legend()
#detalles minimos para la similitud con respecto a la grafica de profesor.
plt.xlim(0, 9)  # Acá limitamos el rango de tiempo para mostrarlo en la grafica resultante
plt.ylim(-0.05, 0.05)  # Acá ponemos los valores limites de la amplitud
plt.xticks(np.arange(0, 9, 2))  # Establecemos las marcas de division del eje x para la similitud
plt.yticks([-0.05, 0, 0.05])  # Establecemos las marcas de division del eje y para la similitud


plt.tight_layout()
plt.show()


```

# 3) Transformaciones de una señal

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del pulso rectangular
amplitud = 1.0
ancho = 0.5
inicio = -1.0
fin = 1.0

# Generar pulso rectangular
tiempo = np.linspace(inicio, fin, num=1000)
pulso_rectangular = np.where((tiempo >= -ancho/2) & (tiempo <= ancho/2), amplitud, 0)

# Reflexiones y transformaciones
reflejada = -pulso_rectangular
desplazada = np.roll(pulso_rectangular, 100)  # Desplazamiento de 100 muestras hacia la derecha
escalada = 0.5 * pulso_rectangular
inversa = np.roll(pulso_rectangular, 100) + pulso_rectangular+2*np.roll(pulso_rectangular, 200)

# Configuración de la figura
fig, axs = plt.subplots(5, 1, figsize=(8, 12))

# Graficar señales
axs[0].plot(tiempo, pulso_rectangular)
axs[0].set_title('Pulso Rectangular')
axs[0].set_xlabel('Tiempo')
axs[0].set_ylabel('Amplitud')
axs[0].set_ylim([-1.5, 1.5])

axs[1].plot(tiempo, reflejada)
axs[1].set_title('Reflejada')
axs[1].set_xlabel('Tiempo')
axs[1].set_ylabel('Amplitud')
axs[1].set_ylim([-1.5, 1.5])

axs[2].plot(tiempo, desplazada)
axs[2].set_title('Desplazada')
axs[2].set_xlabel('Tiempo')
axs[2].set_ylabel('Amplitud')
axs[2].set_ylim([-1.5, 1.5])

axs[3].plot(tiempo, escalada)
axs[3].set_title('Escalada')
axs[3].set_xlabel('Tiempo')
axs[3].set_ylabel('Amplitud')
axs[3].set_ylim([-1.5, 1.5])

axs[4].plot(tiempo, inversa)
axs[4].set_title('Modificada')
axs[4].set_xlabel('Tiempo')
axs[4].set_ylabel('Amplitud')
axs[4].set_ylim([-1.5, 4.5])

plt.tight_layout()
plt.show()
```
